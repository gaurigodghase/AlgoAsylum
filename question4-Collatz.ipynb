### Collatz using Recursion

```python
def collatz(n: int,counter: int) :
    """Collatz Conjecture using recursion"""
    i = n
    if n == 1:
        print("number of iterations is :",counter)
    elif n%2 == 0:
        counter = counter + 1
        collatz(n/2,counter)
    else:
        counter = counter + 1
        collatz(3*n+1,counter)
```

```python

n = int(input("Enter the value whose Collatz conjecture we have to find."))
collatz(n,counter=0)
```

### Collatz without using Recursion

```python
def collatz_using_loop(n:int)->int:
    """Collatz Conjecture using loop"""
    counter = 0
    while (n > 1):
        counter = counter + 1
        if (n%2 == 0):
            
            n=n//2
        else:
            n = 3*n + 1
            
    return (counter)
```

```python
#To find collatz conjecture of a range 
n = int(input("Enter range to find number of iterations"))
collatz_values = []

for i in range (1,n+1):
    collatz_values.append(collatz_using_loop(i))
print(collatz_values)
    
```
