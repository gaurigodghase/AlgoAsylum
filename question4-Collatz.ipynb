### Collatz without using Recursion

```python
def collatz_using_loop(n:int)->int:
    """Collatz Conjecture using loop"""
    counter = 0
    while (n > 1):
        counter = counter + 1
        if (n%2 == 0):    
            n=n//2
        else:
            n = 3*n + 1
            
    return (counter)
```

```python
#To find collatz conjecture of a range 
def my_old_function():
    """Using for Loop and append"""
    n = 5000 #n = int(input("Enter range to find number of iterations"))
    collatz_values = []

    for i in range (1,n+1):
        collatz_values.append(collatz_using_loop(i))
    print(collatz_values)
    
```

```python
def my_new_function():
    """Using List Comprehension"""
    n = 5000 #n = int(input("Enter range to find number of iterations"))
    collatz_values = [ collatz_using_loop(x) for x in range(1,n+1)]
    print(collatz_values)
```
